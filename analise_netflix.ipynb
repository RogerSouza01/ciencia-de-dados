```python
#Importando Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Carregar os dados
df = pd.read_csv('Netflix_Movies_and_TV_Shows.csv')
df.head()
```

```python
#Limpeza de dados
df['Release Year'] = df['Release Year'].astype(int)
df.isnull().sum()
```

```python
#Analise de Generos
genre_counts = df['Genre'].value_counts()
genre_counts.head(10).plot(kind='bar', color='teal')
plt.title('Gêneros Mais Comuns na Netflix')
plt.xlabel('Gênero')
plt.ylabel('Número de Títulos')
plt.show()
```
